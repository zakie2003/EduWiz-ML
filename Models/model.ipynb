```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
data=pd.read_csv("../DataSet/student_exam_data.csv")
```

```python
data
```

```python
question_count = data['Question_ID'].value_counts()
question_count.columns = ['Question_ID', 'count']
data = pd.merge(data, question_count, on='Question_ID', how='left')
data["Accuracy"]=data["Correct"]/data["count"]
```

```python
data["Accuracy"].value_counts()
```

```python
data["count"].value_counts()
```

```python
data["Gender"].unique()
```

```python
data=data.drop(columns=["Question_ID","Student_ID","count"])
data.to_csv("../DataSet/student_api.csv",index=False)
```

```python
data["Difficulty"]=data["Difficulty"].map({"Easy":1,"Medium":2,"Hard":3})
data["Correct"]=data["Correct"]*1
data[data["Gender"].unique()]=pd.get_dummies(data["Gender"])*1
data[data["Learning_Style"].unique()]=pd.get_dummies(data["Learning_Style"])*1
data[data["Subject"].unique()]=pd.get_dummies(data["Subject"])*1
data
```

```python
std=MinMaxScaler()
data[["Age","Time_Taken","Previous_Score"]]=std.fit_transform(data[["Age","Time_Taken","Previous_Score"]])
data=data.drop(columns=["Gender","Learning_Style","Question_Type","Subject"])
```

```python
data
```

```python
temp_data = {
    'Age': [0.6],
    'Previous_Score': [0.87],
    'Difficulty': [3],
    'Correct': [1],
    'Time_Taken': [0.3],
    'Accuracy': [0.9],
    'Male': [1],
    'Female': [0],
    'Reading/Writing': [1],
    'Auditory': [0],
    'Kinesthetic': [0],
    'Visual': [1],
    'Science': [1],
    'Geography': [0],
    'English': [1],
    'History': [0],
    'Math': [1]
}
student_profile = pd.DataFrame(temp_data)
(student_profile)
```

```python
cosine_sim = cosine_similarity(data,student_profile)
cosine_sim_df = pd.DataFrame(cosine_sim)
print(cosine_sim_df)
```

```python
top_similar_students = cosine_sim_df[0].sort_values(ascending=False).index[1:]
top_similar_students
```
